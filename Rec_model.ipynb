# Объединение october_promo и train_target

```python
# Перевод столбцов-дат в нужный формат
clients_promo_october["LocalBeginDate"] = pd.to_datetime(clients_promo_october["LocalBeginDate"]) 
clients_promo_october["LocalEndDate"] = pd.to_datetime(clients_promo_october["LocalEndDate"])

orders["SaleDate"] = pd.to_datetime(orders["SaleDate"])

# Колонки для объединения
orders_cols = ["ClientUUId", "SaleDate", "apply_promo"]

october_promo = clients_promo_october.merge(orders[orders_cols], how="inner", on="ClientUUId")

# Оставляем только промо, подходящих по дате
october_promo = october_promo[
        (october_promo["LocalBeginDate"] <= october_promo["SaleDate"]) & 
        (october_promo["SaleDate"] <= october_promo["LocalEndDate"])
    ].drop_duplicates()

october_promo = october_promo.drop(columns=["SaleDate"])
```

```python
train_target = pd.concat([train_target, october_promo])

train_target
```

```python
unique_orders = orders.drop_duplicates(subset="OrderUUId")

mean_total_order_price = unique_orders.groupby("ClientUUId")["OrderTotalPrice"].mean().reset_index()

mean_total_order_price
```

```python
transformed_orders = transformed_orders.merge(mean_total_order_price, how="left", on="ClientUUId")

transformed_orders = transformed_orders.rename(columns={"OrderTotalPrice": "MeanOrderTotalPrice"})

transformed_orders
```

## Добавление CountProductCategory

```python
product_categories = pd.get_dummies(orders["CategoryId"], prefix="ProductCategory")

orders_with_product_categories = pd.concat([orders, product_categories], axis=1)

orders_with_product_categories
```

```python
count_product_categories = orders_with_product_categories.groupby("ClientUUId")[[
    "ProductCategory_" + str(x) for x in range(1, 8)
]].sum().reset_index() 
```

```python
count_product_categories
```

```python
transformed_orders = transformed_orders.merge(count_product_categories, how="left", on="ClientUUId")

transformed_orders
```

## Добавление CountOrderType

```python
order_types = pd.get_dummies(orders["OrderType"], prefix="OrderType")

orders_with_order_types = pd.concat([orders, order_types], axis=1)

orders_with_order_types
```

```python
count_order_types = orders_with_order_types.groupby("ClientUUId")[[
    "OrderType_" + str(x) for x in range(1, 4)
]].sum().reset_index() 
```

```python
transformed_orders = transformed_orders.merge(count_order_types, how="left", on="ClientUUId")

transformed_orders
```

# Обучение модели (попытка 1)

```python
train_data = train_target.merge(transformed_orders, how="left", on="ClientUUId")

train_data[train_data.isna().any(axis=1)]
```

# Преобразование признака Discount

```python
X
```

```python
# скидки в рублях/купонах переводим в проценты
def convert_discount(promo_id, order_price, discount):
    if promo_id == 6:
        return discount / 100
    
    else:
        return discount / order_price
    
X["Discount"] = X.apply(lambda x: convert_discount(x.Id, x.OrderPrice, x.Discount), axis=1)

X
```

# Обучение модели (попытка 4)

# График корреляции

```python
from scipy.stats import pointbiserialr, pearsonr


correlations = []

for column in X.columns:
    if X[column].nunique() < 6:
        corr, _ = pointbiserialr(X[column], y)
    
    else:
        corr, _ = pearsonr(X[column], y)
    
    correlations.append(corr)

correlation_df = pd.DataFrame(correlations, index=X.columns, columns=['Correlation'])

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_df, annot=True, cmap='coolwarm', center=0)
plt.title('Корреляция между признаками и таргетом')
plt.show()
```

## Анализ дисперсии

```python
from sklearn.decomposition import PCA


rfc_for_features = RandomForestClassifier(random_state=42)

pca = PCA(n_components=None, random_state=42)
```

```python
pca.fit(X_scaled)

explained_variance_cumulative = pca.explained_variance_ratio_.cumsum()

plt.subplot(2,1,2)
plt.plot(list(range(1, 57)),explained_variance_cumulative)
plt.title("Cumulative Variance across Principal Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative % Variance")
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components = 0.95,  random_state = 42)

# fit to our data
X_pca = pca.fit_transform(X_scaled)

print(pca.n_components_)
```

## Adaptive Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier


# Базовый классификатор - дерево решений с ограниченной глубиной
base_estimator = DecisionTreeClassifier()

# Создание модели AdaBoost
adbc = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, learning_rate=1.0, random_state=42)

scores = cross_val_score(
    adbc, X_pca, y, scoring='roc_auc', n_jobs=-1)
# accuracy
print('Accuracy: %.3f ,\nStandard Deviations :%.3f' %
      (np.mean(scores), np.std(scores)))
```

## CatBoosting

```python
from catboost import CatBoostClassifier


class_weights = {0: 1, 1: 6}  # Пример весов для классов 0 и 1

# Создание модели CatBoost
catb = CatBoostClassifier(iterations=500,
                           learning_rate=0.1,
                           class_weights=class_weights)

scores = cross_val_score(
    catb, X_pca, y, scoring='roc_auc', n_jobs=-1)
# accuracy
print('Accuracy: %.3f ,\nStandard Deviations :%.3f' %
      (np.mean(scores), np.std(scores)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42, stratify=y)

class_weights = {0: 1, 1: 6}  # Пример весов для классов 0 и 1

# Создание модели CatBoost
catb = CatBoostClassifier(iterations=500,
                           learning_rate=0.1,
                           class_weights=class_weights)

catb.fit(X_train, y_train)

# Оценить модель на оригинальной тестовой выборке с использованием вероятностей
y_pred_proba = catb.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f'ROC AUC на оригинальных данных: {roc_auc}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score


X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42, stratify=y)

# Применить SMOTE только к обучающей выборке
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Обучить модель на сбалансированной обучающей выборке
model = RandomForestClassifier(random_state=42)
model.fit(X_train_smote, y_train_smote)

# Оценить модель на оригинальной тестовой выборке с использованием вероятностей
y_pred_proba = model.predict_proba(X_test)[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f'ROC AUC на оригинальных данных: {roc_auc}')
```

# Подбор параметров модели

## CatBoosting

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'iterations': [100, 200, 500],
    'learning_rate': [0.01, 0.05, 0.1],
    'depth': [4, 6, 8],
    'l2_leaf_reg': [1, 3, 5],
    'border_count': [32, 64, 128]
}

class_weights = {0: 1, 1: 6}

model = CatBoostClassifier(loss_function='Logloss', class_weights=class_weights, verbose=0, random_state=42)

# Set the gridsearch model
grid = GridSearchCV(estimator = model,
                    param_grid = param_grid, 
                    scoring = 'roc_auc',
                    cv = 3, 
                    n_jobs = -1)
# Fit the gridsearch
grid.fit(X_train, y_train)
```
